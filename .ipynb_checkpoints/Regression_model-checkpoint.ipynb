# Regression model to predict COVID-19 incidence rate per day

### Install packages

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split, GridSearchCV
import matplotlib.pyplot as plt
```

```python
agg_data=pd.read_csv('C:/Users/matth/OneDrive/Documents/Projects/Science/Sentinel/COVID tracker/aggregated_covid_data.csv',sep = '|',\
                    index_col=0)
agg_data.head()
```

### When do these values start?

```python
for col in agg_data.columns:
    print(col, agg_data[col].first_valid_index())
```

```python
for col in agg_data.columns:
    print(col, agg_data[col].last_valid_index())
```

So it looks like I need to drop ER adm, since that doesn't start until October of 2022, and the last reliable testing cases were in Jan 2022. Same with Covid test poitivity rate. I can use hospitalization admission rate, COVID-NET, and the michigan plants. That's about it. 

If that doesn't work, I could also bring in vaccinations and variants, and use all the columns to predict hospitalizations from Aug 2020 to April 2024, and then use hospitalizations to predict actual cases per day.

```python
clean_data=agg_data.drop(['ER_adm','CTK_pos','New_COVID_7DayADM_rate','NWSS_wv_2553_Treatment plant_raw wastewater',\
                                            'NWSS_wi_231_Treatment plant_raw wastewater','NWSS_mi_761_Treatment plant_post grit removal'],axis=1)\
                                            .ffill().bfill()
clean_data['infectious']=clean_data['confirmed_cases'].rolling(13).sum()
comp_data=clean_data.loc['2020-03-25':'2022-01-26']
```

```python
comp_data.head()
```

```python
fig, ax = plt.subplots(3,1, figsize=(16,8))
ax[0].plot(comp_data['infectious'])
ax[1].plot(comp_data['NWSS_mi_757_Treatment plant_post grit removal'])
ax[2].plot(comp_data['COVID-NET'])
fig.autofmt_xdate()
```

### Split cleaned and formatted data

```python
y=comp_data['infectious']
X = comp_data.drop(['infectious','confirmed_cases'], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.33)
```

### Grid Search and Model Tuning

```python
regr_test= GradientBoostingRegressor()
```

```python
param_grid = {'learning_rate' : [ 0.05,0.1,0.25],
              'n_estimators': [500,750, 1000],
              'max_depth': [4],
              'subsample' : [0.66, 0.75,1]}
```

```python
grid_search = GridSearchCV(estimator=regr_test, param_grid=param_grid, scoring= 'neg_mean_squared_error', cv=5, n_jobs=-1)
```

```python
grid_search.fit(X_train, y_train)
```

```python
print("Best Hyperparameters:", grid_search.best_params_)
print("Best Score:", grid_search.best_score_)
```

```python
regr= GradientBoostingRegressor(learning_rate = 0.25, max_depth = 4, n_estimators = 750, subsample = 0.75)
```

```python
regr.fit(X_train, y_train)
```

```python
results = pd.DataFrame( {'predictions' :regr.predict(X_test),
                         'actuals':y_test.values}, index = y_test.index)
results.head()
```

```python
fig, ax = plt.subplots(figsize = (16,8))
ax.plot(results)
fig.autofmt_xdate()
```

```python
full_results = pd.DataFrame( {'predictions' :regr.predict(X),
                         'actuals':y.values}, index = y.index)
full_results.head()
```

```python
fig, ax = plt.subplots(figsize = (16,8))
ax.plot(full_results)
fig.autofmt_xdate()
```
